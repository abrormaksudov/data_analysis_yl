        Проверка гипотез

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np


data = pd.read_csv('data_part_2.csv')
```

1) Влияет ли тип устройства на количество покупок в день по каждому региону?

```python
#Подготовка
purchases = data[data['sold'] == 1]
purchases['order_dt'] = pd.to_datetime(purchases['order_dt'])
purchases['purchase_date'] = purchases['order_dt'].dt.date
grouped = purchases.groupby(['region', 'device', 'purchase_date']).size().reset_index(name='purchase_count')

#1. Визуализция
plt.figure(figsize=(12, 6))
sns.boxplot(x='device', y='purchase_count', hue='region', data=grouped)
plt.title('Boxplot: Количество покупок по устройствам и регионам')
plt.show()
plt.figure(figsize=(12, 6))
sns.kdeplot(data=grouped, x='purchase_count', hue='device', fill=True)
plt.title('KDE Plot: Распределение количества покупок по устройствам')
plt.show()
#4. Проверка на нормальность
print('p-value:')
for device in grouped['device'].unique():
    device_data = grouped[grouped['device'] == device]['purchase_count']
    stat, p_value = stats.shapiro(device_data)
    print(f'    {device} - {p_value}')
print('\nНормальность -', all(stats.shapiro(grouped[grouped['device'] == device]['purchase_count'])[1] > 0.05 for device in grouped['device'].unique()))
#6. Задаём альфа и применяем стат тест
test_type = 'Kruskal-Wallis'
alpha = 0.05
h_stat, p_value = stats.kruskal(*[grouped[grouped['device'] == device]['purchase_count'] for device in grouped['device'].unique()])
print(f'\nконечное p-value: {p_value}')
print('Вывод: Тип устройства не влияет на количество покупок в день по регионам.')
```

2. Влияет ли тип рекламного канала на количество покупок в день по каждому региону?

```python
#Подготовка
grouped = purchases.groupby(['region', 'channel', 'purchase_date']).size().reset_index(name='purchase_count')

#1. Визуализция
plt.figure(figsize=(12, 6))
sns.boxplot(x='channel', y='purchase_count', hue='region', data=grouped)
plt.title('Boxplot: Количество покупок по рекламным каналам и регионам')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(12, 6))
sns.kdeplot(data=grouped, x='purchase_count', hue='channel', fill=True)
plt.title('KDE Plot: Распределение количества покупок по рекламным каналам')
plt.show()

#4. Проверка на нормальность
print('p-value:')
for channel in grouped['channel'].unique():
    channel_data = grouped[grouped['channel'] == channel]['purchase_count']
    stat, p_value = stats.shapiro(channel_data)
    print(f'    {channel} - {p_value}')
print('\nНормальность -', all(stats.shapiro(purchases[purchases['hour'] == hour]['final_price'])[1] > 0.05 for hour in purchases['hour'].unique()))

#6. Задаём альфа и применяем стат тест
test_type = 'Kruskal-Wallis'
alpha = 0.05
h_stat, p_value = stats.kruskal(*[grouped[grouped['channel'] == channel]['purchase_count'] for channel in grouped['channel'].unique()])
print(f'\np-value: {p_value}')
print('Вывод: Тип рекламного канала влияет на количество покупок в день по регионам.')
```

5. Проверить гипотезу о том, что средний чек отличается в зависимости от времени суток?

```python
#Подготовка
purchases['session_start'] = pd.to_datetime(purchases['session_start'])
purchases['hour'] = purchases['session_start'].dt.hour

#1. Визуализция
plt.figure(figsize=(12, 6))
sns.boxplot(x='hour', y='final_price', data=purchases)
plt.title('Boxplot: Средний чек по часам')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(12, 6))
sns.kdeplot(data=purchases, x='final_price', hue='hour', fill=True)
plt.title('KDE Plot: Распределение среднего чека по часам')
plt.show()

#4. Проверка на нормальность
print('p-value:')
for hour in purchases['hour'].unique():
    hour_data = purchases[purchases['hour'] == hour]['final_price']
    stat, p_value = stats.shapiro(hour_data)
    print(f'    {hour}, p-value: {p_value}')
print('\nНормальность -', all(stats.shapiro(purchases[purchases['hour'] == hour]['final_price'])[1] > 0.05 for hour in purchases['hour'].unique()))

#6. Задаём альфа и применяем стат тест
test_type = 'Kruskal-Wallis'
alpha = 0.05
h_stat, p_value = stats.kruskal(*[purchases[purchases['hour'] == hour]['final_price'] for hour in purchases['hour'].unique()])
print(f'\np-value: {p_value}')
print('Вывод: Средний чек не отличается в зависимости от рекламного канала.')
```
