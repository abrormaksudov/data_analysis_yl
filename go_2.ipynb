```python
df['week'] = df['session_date'].dt.isocalendar().week
df['week'] = df['session_date'].dt.strftime('%m-W%V')
df['week']
```

```python
def plot_month(month, hue, rev=False):
    usa_df = df[(df['region'] == 'United States') & (df['month'] == month)]
    france_df = df[(df['region'] == 'France') & (df['month'] == month)]
    germany_df = df[(df['region'] == 'Germany') & (df['month'] == month)]
    uk_df = df[(df['region'] == 'UK') & (df['month'] == month)]

    
    
    dfs = [(usa_df, [0, 0]), (france_df, [0, 1]), (germany_df, [1, 0]), (uk_df, [1, 1])]
    
    fig, ax = plt.subplots(2, 2, figsize=(12, 3.5))
    total = 0
    total_prop = 0
    if hue == 'revenue' or rev:
        df_copy = df.query('revenue > 0')
    else:
        df_copy = df
    for df_go, (x, y) in dfs:
        region = df_go['region'].unique()[0]
        if hue == 'revenue' or rev:
            df_go = df_go.query('revenue > 0')
            df_copy = df.query('revenue > 0')
        group = df_go.groupby('week')[hue].count()
        
        total += group.sum()
        
        total_prop += round(group.sum() / len(df_copy) * 100, 2)
        # print(group.sum(), round(group.sum() / len(df_copy) * 100, 2))
        # print(region, group)
        if hue == 'revenue' or rev:
            df_go = df_go.query('revenue > 0')
        grouped = df_go.groupby(['week', hue], observed=True).size().reset_index(name='count')

        sns.barplot(data=grouped, x='week', y='count', hue=hue, hue_order=grouped[hue].unique(), ax=ax[x,y])
        ax[x,y].set_xlabel('')
        ax[x,y].set_ylabel('')
        ax[x,y].set_title(region)
        ax[x,y].legend(loc='best', fontsize=10)
    print("total:", total, round(total_prop, 2))
    plt.show()
```

```python
for i in range(5, 11): plot_month(i, 'revenue')
```

```python
for i in range(5, 11): plot_month(i, 'device', rev=True)
```

```python
for i in range(5, 11): plot_month(i, 'channel')
```
